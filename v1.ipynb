# Visualización 1: Bar chart

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Dataset: https://www.kaggle.com/datasets/andrewmvd/tsunami-dataset
df = pd.read_csv("tsunami_dataset.csv")
mag6 = df[(df.YEAR >= 1900) & (round(df.EQ_MAGNITUDE) == 6)]
mag7 = df[(df.YEAR >= 1900) & (round(df.EQ_MAGNITUDE) == 7)]
mag8 = df[(df.YEAR >= 1900) & (round(df.EQ_MAGNITUDE) == 8)]
mag9 = df[(df.YEAR >= 1900) & (round(df.EQ_MAGNITUDE) == 9)]
```

```python
plt.bar(['Magnitud 6', 'Magnitud 7', 'Magnitud 8', 'Magnitud 9'], [len(mag6), len(mag7), len(mag8), len(mag9)])
plt.title('Número de tsunamis')
```
