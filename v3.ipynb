# Visualización 3: Grid map

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Dataset: https://www.kaggle.com/datasets/andrewmvd/tsunami-dataset
df = pd.read_csv("tsunami_dataset.csv")
df = df[(df.YEAR >= 1900) & (df.EQ_MAGNITUDE > 6)]
df = df[["LATITUDE", "LONGITUDE", "EQ_MAGNITUDE"]]
df
```

```python
countries = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
countries.head()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
countries.plot(color="lightgrey", ax=ax)
df.plot(x="LONGITUDE", y="LATITUDE", kind="scatter", c="EQ_MAGNITUDE", colormap="YlOrRd", ax=ax)
ax.grid(b=True, alpha=0.5)
plt.show()
```
