# Visualizaci√≥n 2: Flow map

```python
import schemdraw
from schemdraw import flow
```

```python
with schemdraw.Drawing() as d:
    d.config(fontsize=10, unit=.5)
    d += flow.Terminal().label('Inicio')
    d += flow.Arrow()
    d += flow.Process().label('Buscar banco').drop('E')
    d += flow.Arrow().right()
    d += flow.Process().label('Reclutar\ncolaboradores').drop('E')
    d += flow.Arrow().right()
    d += flow.Process().label('Cavar t√∫nel').drop('S')
    d += flow.Arrow().down()
    d += (d1 := flow.Decision(E='S√≠', S='No').label('¬øEres del grupo\nde los atracadores?'))
    
    d += flow.Arrow().at(d1.E).right(d.unit * 2)
    d += flow.Process().label('Fingir robo expr√©s').drop('S')
    d += flow.Arrow().down(d.unit * 12)
    
    d += flow.Arrow().at(d1.S).down(d.unit * 2)
    d += flow.Process().label('Esperar en el t√∫nel')
    d += flow.Arrow().down()
    d += flow.Process().label('Taladrar pared')
    d += flow.Arrow().down()
    d += flow.Process().label('Abrir cajas').drop('E')
    d += flow.Arrow().right(d.unit * 3)
    d += flow.Process().label('Reunirse todos').drop('S')
    d += flow.Arrow().down(d.unit * 2)
    d += flow.Process().label('Salir por el t√∫nel').drop('S')
    d += flow.Arrow().down(d.unit * 2)
    d += flow.Terminal().label('Fin')
    d.save('flow-map.jpg')
```
